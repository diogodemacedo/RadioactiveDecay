The aim of this notebook is to use the process of radioactive decay as an example of how to solve first order ordinary differential equations (ODE's) by applying the Euler method.

<br /> 

### Background Theory
<br /> 

**Radioactive decay** is the process by which an unstable atomic nucleus loses energy by emitting radiation in the form of particles or electromagnetic waves. This physical process occurs in isotopes that have an imbalance of protons and neutrons in their nucleus, making it unstable. As a consequence, the unstable nucleus udnergoes a transformation into a different nucleus with a lower energy state and a more stable configuration. 
There are three types of radioactive decay:

* **Alpha-Decay**: An $\alpha$ particle, which consists of two protons and two neutrons, is emitted from the nucleus.
* **Beta-Decay**: A neutron in the nucleus is converted into a proton and an electron. The electron is consequently emitted from the nucleus int the form of a $\beta$ particle.
* **Gamma-Decay**: The nucleus releases excess energy in the form of high-energy photons, referred to as $\gamma$-rays.

Radioactive decay plays a fundamental role in many natural processes, such as nuclear energy production and radioactive dating. The main cons associated with this process are the possible health hazards that occur if the emitted radiation is not properly controlled or shielded.

Let us now express the radioactive decay process mathematically. Given a sample of a particular radioisotope, the number of decay events, $âˆ’dN$, expected to occur in an infinitesimally small time interval, $dt$, is proportional to the number of atoms present, $N$, that is:

\begin{equation}
-\frac{dN}{dt} \propto N
\end{equation}

Note that the negative sign simply indicates that $N$ is decreasing as time increases. Introducing now the decay proportionality constant, $\lambda$, which depends on the radioactive isotope and determines the rate at which it decays, it comes that:

\begin{equation}
-\frac{dN}{dt} = \lambda N
\end{equation}

Solving this first order ordinary differential equation involves a technique known as *separation of variables*, which will be used below.
<br /> 


#### Analytical Solution

The separation of variables method consists of rewriting the differential equation in such a way that each of the two variables appears on a different side of the equation. Applying this to the equation given above yields:

\begin{equation}
-\frac{dN}{N}=\lambda dt
\end{equation}

Integrating both sides,

\begin{equation}
    \int_{N_0}^{N} -\frac{dN}{N} = \int_{t_0}^{t} \lambda dt
\end{equation}

Solving this integral, and specifying initial conditions as $t_0 = 0$ and $N_0 = N_0$, we can establish that:

\begin{equation}
    \ln(N)-\ln(N_0) = - \lambda t
\end{equation}

Raising $e$ to both sides,

\begin{equation}
    e^{\ln\big(\frac{N}{N_0}\big)} = e^{-\lambda t}
\end{equation}

And lastly, solving for $N$:

\begin{equation}
    N(t) = N_0 e^{- \lambda t}
\end{equation}

Now that the analytical solution has been derived, we can proceed with a possible numerical method employed to solve this type of equations.

#### Numerical Solution

<br /> 

The numerical technique I will program to solve this equation is called *Euler Method*. In  a general abstract form, a first order ODE can be written as: 

\begin{equation}
\frac{dy}{dt}= f(t,y)
\end{equation}

Recalling the approximate definition of derivative, we can write the previous equation as:

\begin{equation}
\frac{dy}{dt} = \frac{y(t+h)-y(t)}{h} \Rightarrow y(t+h) = \frac{dy}{dt}h + y(t)
\end{equation}

Where $h$ represents the time interval between successive evaluations of the function. In computer programming, $h$ is usually refereed to as *time step*.
